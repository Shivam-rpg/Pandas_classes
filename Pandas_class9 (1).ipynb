## 1. Applying Custom Functions with `apply()` / `map()`
- `apply()` lets you apply a custom function across rows or columns.
- `map()` is for applying functions element-wise to a Series.
- Both are handy for transformations, calculations, and cleaning data.

```python
import pandas as pd
import numpy as np
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Emma"],
    "Math": [88, 75, 92, 60, 85],
    "Science": [90, 80, 78, 70, 95]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
df['Total_Sum'] = df.apply(lambda row: row['Math'] + row['Science'], axis=1)

#or
df['Total_Sum'] = df[['Math', 'Science']].apply(lambda row: row.sum(), axis = 1)
df
```

```python
def to_grade(s):
  if s>= 90:
    return "A"
  elif s >= 80:
    return "B"
  elif s >= 70:
    return "C"
  else:
    return "D"
df["Math_grade"] = df["Math"].apply(to_grade)
```

```python
df
```

```python
df['Name'] = df['Name'].map(str.lower)
df
```

```python
df["Bonus Marks"] = df['Math'].map(lambda x: x+5)
df
```
